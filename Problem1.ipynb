```python
# Import delay data
Delay_cause = pd.read_csv('Delay_cause.csv')

# Create a variable - TOTAL_DELAY
Delay_cause['TOTAL_DELAY'] = Delay_cause['CARRIER_DELAY'] + Delay_cause['WEATHER_DELAY'] + Delay_cause['NAS_DELAY'] + Delay_cause['SECURITY_DELAY'] + Delay_cause['LATE_AIRCRAFT_DELAY']

# Create a variable - DELAY, indicate the flight is delay or not (Delay = Any Delay above 0 minutes)
DELAY = np.zeros(np.size(Delay_cause['TOTAL_DELAY']))
Delay = np.where(Delay_cause['TOTAL_DELAY'] >= 0)
DELAY[Delay] = 1
Delay_cause['DELAY'] = DELAY

Delay_cause.head()
```
